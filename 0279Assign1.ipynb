```python
#TASK 3:
def primes(n):
    if n < 2:
        return
    marked = [0] * (n+1);
    value = 3
    yield 2
    while value <= n:
        if marked[value] == 0:
            yield value
            i = value
            while i <= n:
                marked[i] = 1
                i += value
        value += 2

def is_lr_truncatable(n, P):
    if n not in P:
        return False
    s = str(n)[1:]
    if len(s) == 0:
        return True
    return is_lr_truncatable(int(s), P)

def is_rl_truncatable(n, P):
    if n not in P:
        return False
    s = str(n)[:-1]
    if len(s) == 0:
        return True
    return is_rl_truncatable(int(s), P)

def is_truncatable_prime(n, P):
    if n not in P:
        return False
    if n < 10:
        return False
    return is_leftright_truncatable(n,P) and is_rl_truncatable(n,P)

def truncatable_primes(limit):
    P = set(primes(limit))
    for i in P:
        if is_truncatable_prime(i, P):
            yield i

A = list(truncatable_primes(1000000))

for p in range(0,len(A)):
    print(A[p])

print("Result is :",sum(A))
```
